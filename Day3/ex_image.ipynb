```python
from PIL import Image
img = Image.open("../Data/image.jpeg").convert("L") # convert to black and white
```

```python
img
```

```python
width, height = img.size
pixel_list = list()
# Iterate through pixels using nested loops
for y in range(height):
    pixel_row = list()
    for x in range(width):
        pixel_value = img.getpixel((x, y))
        pixel_row.append(pixel_value)
    pixel_list.append(pixel_row)
```
